```python
with open('KeyTimeA.txt', 'r') as file:
    lines = file.readlines()
squares = []
for line in lines:
    data = line.strip().split(',')
    if len(data) == 2:
        number = float(data[0])
        square = number ** 2  # Calculate the square
        squares.append(square)   
mean_X2 = sum(squares) / len(squares)
print(f"E[X^2] : {mean_X2}")
```

```python
with open('KeyTimeB.txt', 'r') as file:
    lines = file.readlines()
squares = []
for line in lines:
    data = line.strip().split(',')
    if len(data) == 2:
        number = float(data[0])
        square = number ** 2  # Calculate the square
        squares.append(square)   
mean_Y2 = sum(squares) / len(squares)
print(f"E[Y^2] : {mean_Y2}")
```

```python
var_X = mean_X2 - (mean_X ** 2)
var_Y = mean_Y2 - (mean_Y ** 2)

std_dev_X = np.sqrt(var_X)
std_dev_Y = np.sqrt(var_Y)

x_values = np.linspace(mean_X - 3*std_dev_X, mean_X + 3*std_dev_X, 1000)
y_values = np.linspace(mean_Y - 3*std_dev_Y, mean_Y + 3*std_dev_Y, 1000)

pdf_x = (1/(std_dev_X * np.sqrt(2 * np.pi))) * np.exp(-(x_values - mean_X)**2 / (2 * std_dev_X**2))
pdf_y = (1/(std_dev_Y * np.sqrt(2 * np.pi))) * np.exp(-(y_values - mean_Y)**2 / (2 * std_dev_Y**2))

plt.figure(figsize=(8, 6))
plt.plot(x_values, pdf_x, color='blue', label='X Distribution')
plt.plot(y_values, pdf_y, color='red', label='Y Distribution')

plt.title('Approximated Normal Distributions')
plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)

plt.show()
```

```python
from tabulate import tabulate
import csv
from itertools import count
import operator
import math
import numpy as np

EPSILON = 0.000001

def main():
    

    KeyTimeA_word_prob = make_word_prob_map('KeyTimeA.txt')
    KeyTimeB_word_prob = make_word_prob_map('KeyTimeB.txt')

    

    # Get the word count of the unknown document
    
    unknown_doc_count, n_words = make_word_count_map('email.txt')

    
    print("n_words", n_words)

    KeyTimeA = calc_log_pr_doc_given_author(KeyTimeA_word_prob, unknown_doc_count)
    KeyTimeB = calc_log_pr_doc_given_author(KeyTimeB_word_prob, unknown_doc_count)
    print("log P(E|A)\t", KeyTimeA)
    print("log P(E|B)\t",KeyTimeB)

    print('diff\t',KeyTimeA - KeyTimeB )

def calc_log_pr_doc_given_author(prob_map, counts):
    """
    How likely is the document, given the counts of words in the doc
    and the authors prob_map
    """
    data = []
    log_prob = math.log(1)
    for word_i, c_i in counts.items():
        p_i = get_word_prob(prob_map, word_i)
        log_prob += c_i * math.log(p_i)
        lst = [word_i, c_i, log_prob]
        data.append(lst)
       
    print(tabulate(data))
    return log_prob


def calcLogProbDoc(wordProbMap, countMap):
    logProb = math.log(1)
    for wordi in countMap:
        ci = countMap[wordi]
        pi = get_word_prob(wordProbMap, wordi)
        logProb += ci * math.log(pi)
    return logProb

# If a word is in a probability dictionary, return its probability
# otherwise, return epsilon
def get_word_prob(word_prob_map, word):
    if word in word_prob_map:
        return word_prob_map[word]
    return EPSILON

# From a file name, approximate the probability of a word
# being generated from the same distribution as the file.
# Assume that each word is produced independently, regardless
# of order.
def make_word_prob_map(fileName):
    wordMap, nWords = make_word_count_map(fileName)
    probabilityMap = {}
    for word in wordMap:
        count = wordMap[word]
        p = float(count) / nWords
        probabilityMap[word] = p
    #print(probabilityMap)
    return probabilityMap

# From a file name, count the number of times each word exists
# in that file. Return the result as a map (aka a dictionary)
def make_word_count_map(fileName):
    wordMap = {}
    nWords = 0
    data1 = []
 
    delta_t =extract_deltas_from_text(fileName)
    print(delta_t)
    for word in delta_t:
        add_word_to_count_map(wordMap, word)
        nWords+= 1
                
    return wordMap, nWords

# Add a word to a count map. Makes sure not to crash if the
# word has not been seen before.
def add_word_to_count_map(wordMap, word):
    if not word in wordMap:
        wordMap[word] = 0
    wordMap[word] += 1
    

def extract_deltas_from_text(file_path):
    data = np.genfromtxt(file_path, delimiter=',', dtype='float')
    delta_t = np.diff(data[:, 0])
    return delta_t


if __name__ == '__main__':
    main()
```
