```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

data_A = np.genfromtxt("personKeyTimingA.txt", delimiter=',')
data_B = np.genfromtxt("personKeyTimingB.txt", delimiter=',')
data_email = np.genfromtxt("email.txt", delimiter=',')

# Extracting time values
time_A = data_A[:, 0]
time_B = data_B[:, 0]
time_email = data_email[:, 0]

# Calculate means
mean_A = np.mean(time_A)
mean_B = np.mean(time_B)

# Calculate covariance matrices
cov_A = np.cov(time_A, rowvar=False, ddof=0)
cov_B = np.cov(time_B, rowvar=False, ddof=0)

# Plot Normal distributions
x = np.linspace(mean_A - 3 * np.sqrt(cov_A), mean_A + 3 * np.sqrt(cov_A), 100)
y_X = multivariate_normal.pdf(x, mean=mean_A, cov=cov_A)
plt.plot(x, y_X, label="User A")

y = np.linspace(mean_B - 3 * np.sqrt(cov_B), mean_B + 3 * np.sqrt(cov_B), 100)
y_Y = multivariate_normal.pdf(y, mean=mean_B, cov=cov_B)
plt.plot(y, y_Y, label="User B")

plt.xlabel("Time (ms)")
plt.ylabel("Probability density")
plt.legend()
plt.show()

# Calculate likelihoods for the email data given user A and B
likelihood_A = multivariate_normal.pdf(time_email, mean=mean_A, cov=cov_A)
likelihood_B = multivariate_normal.pdf(time_email, mean=mean_B, cov=cov_B)

# Calculate the ratio of probabilities
prior_A = 0.5
prior_B = 0.5
ratio = (prior_A * likelihood_A) / (prior_B * likelihood_B)
print("The ratio of the probability that user A wrote the email over the probability that user B wrote the email is", ratio)
```
